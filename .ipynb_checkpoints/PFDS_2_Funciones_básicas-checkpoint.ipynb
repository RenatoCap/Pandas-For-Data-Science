Para conocer las funcionalidades básicas de pandas, haremos un análisis de un dataset conocido en el mundo del Machine Learning y el cual podemos encontrar en la página de <a href="https://www.kaggle.com/c/titanic">kaggle</a>. Este dataset contiene los datos de los pasajeros que abordaron el Titanic. En este notebook solo nos limitaremos a hacer analisis de los datos usando algunos métodos y atributos comunes al momento de analizar DataFrames. 

Primero, hay que tener como conocimiento previo a utilizar cualquier dataset o realizar cualquier análisis que todos los datasets están compuestos por filas que conitenen los ejemplos, para nuestro caso serían los pasajeros, y columnas en las que se colocan las features de los ejemplos, para nuestro caso los datos de los pasajeros que abordaron el titanic. 

Como vimos en el notebook anterior, primero importaremos los datos desde nuestra carpeta Dataset mediante el método *read_csv*

```python
import pandas as pd 
```

```python
df_titanic = pd.read_csv('./Datasets/Titanic.csv')
df_titanic
```

En Data Science se suele trabajar con datasets muy gandes, por ello lo más conveniente para conocer con que datos nos encontramos trabajando es utilizar el método head que nos muestra las n primeras filas del DataFrame.

<code>DataFrame.head(n=5)</code>

Como vemos en la sintaxis, este método tiene un parámetro n que indica cuantas filas vamos a querer ver, por defecto, este valor es 5, sin embargo podemos modificarlo y se selccionar cuantas filas queremos ver.

```python
df_titanic.head(10)
```

Asi mismo tambien se puede utilizar el método tail para visualizar las n ultimas filas

```python
df_titanic.tail(10)
```

Sin embargo, hay veces en que no encontramos datos relevantes ni en las primeras filas ni en las últimas filas, por ello existe tambien el método sample que nos permite visualizar filas elegidas de manera aleatoria y poder hacer algunas hipotesis acerca de nuestros datos.

```python
df_titanic.sample(10)
```

Para conocer el tamaño del dataset con el que nos encontramos trabajando usamos el atributo shape que nos da como resultado una tupla con el número de filas y columnas de nuestro dataset.

```python
df_titanic.shape
```

```python
pd.options.display.float_format='{:,.3f}'.format
```

Otro método muy util es describe, el cual nos entrega medidas de tendencia central, asi como tambien medidas de dispersión de las features (columnas) con las que nos encontramos trabajando.

```python
df_titanic.describe()
```

El método describe solo realiza el analisis para datos de tipo númerico, es decir entero o flotante, sin embargo si queremos conocer otros datos como la cantidad, valores unicos, la frecuencia y el valor top. podemos incluir el parametro 'all', tal como se muestra acontinuación.

```python
df_titanic.describe(include='all')
```

Otro método de los DataFrame que nos es de mucha utilidad al momento de hacer el analisis de nuestros datos es info, el cual nos entre las features con su cantidad de datos no nulos y el tipo de datos de estos, tal como podemos ver a continuación.

```python
df_titanic.info()
```

Para solo conocer los tipos de datos de las features con las que estamos trabajando, hacemos lo siguiente.

```python
df_titanic.dtypes
```

Como podemos apreciar, vemos que hay features, como Name, Sex, Age, Ticket, Cabin y Embarked que se encuentran identificadass como tipo objeto, si queremos converit estos tipos objetos a sring, hacemos, los siguiente.

```python
df_titanic.convert_dtypes().dtypes
```

Esta conversión no es la más correcta ya que existen datos que pueden ser guardados como tipos de datos categoricos, para esto tenemos que realizar un analisis de los valores unicos de cada feature, esto lo podemos realizar con el metodo *nunique* .

```python
df_titanic.nunique()
```

Podemos ver del resultado que survived y sex tienen 2 valores unicos; Embarket y  Pclass, 3. Por ende, convertimos los datos de estas features en tipo categorial de la siguiente manera.

```python
df_titanic[['Survived', 'Pclass', 'Sex', 'Embarked']] = df_titanic[['Survived', 'Pclass', 'Sex', 'Embarked']].astype('category')
```

```python
df_titanic['Sex'].unique()
```
